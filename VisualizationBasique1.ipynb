<h2 align=center>L'importance des graphiques</h2>
<br>
La visualisation de données est la présentation de données avec des graphiques. C'est un moyen de résumer vos conclusions et de les afficher sous une forme qui facilite l'interprétation et peut aider à identifier des modèles ou des tendances.
Avoir de superbes visualisations de données rendra votre travail plus intéressant et plus clair.
Dans ce cours, vous apprendrez à utiliser la bibliothèque ggplot2 pour créer de superbes graphes et diagrammes, en personnalisant leur apparence et leur convivialité à votre guise.

### La différence entre les bibliothèques R<a id="ref1"></a>

Les différences entre la bibliothèque de base plot(), fournie avec R, et ggplot2 sont nombreuses.
Ggplot2 a été créé pour répondre aux demandes de conception et était basé sur le livre "The Grammar of Graphics", un livre qui décrit les bases du traçage de données.

Regardons les différences entre eux à l'aide d'un exemple simple.

### plot
La bibliothèque `plot` est la bibliothèque R par défaut pour tracer des graphiques. C'est très simpliste à la fois dans la syntaxe et l'esthétique. Pour l'utiliser pour créer un bar plot, vous utilisez la fonction `barplot`, comme ceci :

```r
count <- table(mtcars$cyl)
barplot(count)
```

### ggplot2
`ggplot2`, comme mentionné ci-dessus, est une bibliothèque spécialisée conçue pour créer des visualisations de données visuellement agréables.
Avant de pouvoir utiliser `ggplot2`, nous devons l'importer dans l'environnement R.
La cellule de code ci-dessous vérifiera si votre système a déjà ggplot, afin de ne pas exécuter `install.packages` sans raison. Ensuite, en utilisant la fonction `library`, nous pouvons alors importer `ggplot2`.

```r
if("ggplot2" %in% rownames(installed.packages()) == FALSE) {install.packages("ggplot2")}
library(ggplot2)
```

Now, let's plot our graph. To plot a simple bar graph using `ggplot2`'s `qplot`, we use this:

```r
qplot(mtcars$cyl, geom = "bar")
```

Ne vous inquiétez pas - nous allons revenir en arrière et apprendre ces méthodes de traçage et d'autres au cours de nos leçons.

Comme vous pouvez le voir, `ggplot2` nous offre un graphique plus joli, mais a une syntaxe légèrement plus complexe que la bibliothèque `plot` par défaut.

#### ggplot2 a deux fonctions principales, `qplot()` et `ggplot()`.

   • `qplot()` offre une syntaxe plus simple similaire à la fonction `plot` par défaut, mais sa personnalisation est limitée.

   • `ggplot()` est la fonction à part entière. Il a beaucoup plus de personnalisations possibles, mais a une syntaxe plus compliquée que `qplot()`.

Dans ce cours, nous commencerons à utiliser `qplot()` puis passerons à `ggplot()` au fur et à mesure que vous avancez.

Le graphique suivant est une démonstration de ce qu'il est possible de faire avec `ggplot()`. À la fin de ce cours, vous serez capable de créer des graphiques comme celui-ci ci-dessous et des graphiques encore plus complexes. 

```r
ggplot(data= mtcars, aes(hp))  + 
  geom_histogram(breaks=seq(40, 350, by = 25),  
                 colour = I("black"),
                 aes(fill=..count..))
```

### Données qualitatives vs quantitatives<a id="ref2"></a>
Une chose à toujours garder à l'esprit est le type de données pour lesquelles vous essayez de créer des graphiques. En général, nous classons les données en deux grands groupes : les données qualitatives (également appelées données catégorielles) et les données quantitatives (également appelées données numériques).

Ce que nous appelons données **qualitatives ou catégorielles** sont des données qui font référence, comme le montre leur nom, à des catégories. Cela peut inclure des données pour les questions « Oui ou Non », il peut s'agir du nom d'un lieu, de la saveur de crème glacée préférée d'une personne ou d'autre chose. C'est quelque chose de très différent des données **quantitatives ou numériques**. Les données quantitatives sont des données qui sont, tout simplement, des nombres. Il s'agit normalement d'une mesure quelconque et peut être manipulé à l'aide de calculs simples.

Les méthodes de traçage et les meilleurs types de graphiques pour chaque type de données sont différents - choisir le meilleur vous aidera grandement à créer des graphiques visuellement agréables.

<h2 align=center>Le jeu de données mtcars</h2>
<br>
« mtcars » est un ensemble de données intégré qui contient des données du magazine américain Motor Trend de 1974, et contient la consommation de carburant et 10 autres aspects des performances automobiles pour 32 automobiles de 1973 à 1974.

Comme il est déjà inclus avec R, il n'est pas nécessaire d'importer `mtcars`. Vérifions sa structure - nous pouvons le faire en l'appelant par son nom, comme ceci :

```r
mtcars
```

Quels types d'informations pouvons-nous tirer de ces données ?
Nous avons le kilométrage des voitures par gallon d'essence, le nombre de cylindres et d'autres attributs.

S'il y a une colonne dont vous ne savez pas ce qu'elle représente, vous pouvez utiliser la fonction `?` pour voir le fichier d'aide pour cet ensemble de données, comme ceci :

```r
?mtcars
```

Notez que cela n'est possible que parce que `mtcars` est un jeu de données intégré dans R.
Presque aucun ensemble de données n'aura de fichiers d'aide.
Certains ensembles de données que vous trouverez sur Internet auront un fichier readme qui décrira ce que chaque colonne représente. Certains incluront le nom de chaque colonne de l'ensemble de données. Bien qu'il s'agisse de bonnes pratiques, vous pouvez trouver des jeux de données avec des noms peu descriptifs pour ses colonnes et des jeux de données sans aucun nom.

Comme vous le savez peut-être déjà, nous n'avons pas besoin d'afficher l'ensemble de données dans son intégralité pour voir comment les données sont structurées. Nous pouvons utiliser la fonction `head` pour voir les 6 premières lignes de nos données.

```r
head(mtcars)
```

Et nous pouvons utiliser `tail` pour voir les 6 dernières lignes de nos données.

```r
tail(mtcars)
```

Analysons plus avant nos données...
Nous pouvons obtenir un résumé rapide de chaque colonne en utilisant la fonction « résumé » :

```r
summary(mtcars)
```

Nous pouvons obtenir la moyenne pour n'importe quelle colonne en utilisant `mean(datasetname$columnname)`, comme ceci :

```r
mean(mtcars$cyl)
```

<h2 align=center>Créer des diagrammes à barres</h2>

Commençons notre traçage avec des diagrammes à barres. Comme son nom l'indique, il s'agit d'un format de tracé qui affiche vos données à l'aide de barres.
Vous en avez probablement déjà vu beaucoup.

Avant de créer un diagramme à barres, importons notre bibliothèque de traçage, `ggplot2`.
Il n'est pas nécessaire d'exécuter ce bloc si vous avez déjà exécuté l'import dans "Différences entre bibliothèques R".

```r
if("ggplot2" %in% rownames(installed.packages()) == FALSE) {install.packages("ggplot2")}
library(ggplot2)
```

Maintenant que nous avons chargé nos bibliothèques, commençons à tracer. Pour tracer facilement en utilisant `ggplot2`, nous pouvons utiliser la fonction `qplot`, qui a une syntaxe plus simple, comme ceci :

```r
qplot(mtcars$cyl, geom = "bar")
```

Comme nous pouvons le voir, nous avons tracé un digramme à barres, composé du nombre de chaque élément avec la même valeur.
Nous pouvons maintenant exploiter certaines des possibilités de la fonction `qplot` de `ggplot2`.

```r
?qplot
```

Vous n'avez pas besoin de comprendre comment fonctionne exactement chaque paramètre de la fonction `qplot`, mais vous pouvez toujours consulter le fichier d'aide pour essayer de trouver ce dont vous avez besoin.

Notre graphique est simple tel qu'il se présente actuellement. Un graphique simple est un excellent choix pour les articles universitaires, mais pour le contenu Internet… le simple fait d'être gris n'attirera pas l'attention des gens.
Donnons-lui de la couleur en utilisant les paramètres `color` et `fill`.
`color` modifiera la couleur du contour, tandis que `fill` changera la couleur des barres.

```r
qplot(mtcars$cyl, geom = "bar", fill = I("black"), colour = I("red"))
```

Nous pouvons simplement changer le remplissage, car le contour, si aucun paramètre n'est passé, recevra la même valeur que le remplissage.
Vous pouvez changer les couleurs en bleu, rose, vert, jaune...
Et il y a plus de couleurs !
Voici une liste : http://sape.inf.usi.ch/quick-reference/ggplot2/colour

Nous pouvons également changer le nom de nos axes pour rendre le graphique plus compréhensible en passant les paramètres `xlab` et `ylab` (`lab` signifie "label") :

```r
qplot(mtcars$cyl, geom = "bar", fill = I("blue"), xlab = "Cylinders", ylab = "Number of Vehicles")
```

Pour terminer notre diagramme à barres, nous pouvons donner un nom à notre graphique. Nous pouvons le faire en utilisant le paramètre `main`.

```r
qplot(mtcars$cyl, geom = "bar", fill = I("blue"), xlab = "Cylinders", ylab = "Number of Vehicles", main = "Cylinders in mtcars")
```

<h2 align=center>Histogrammes</h2>
<br>
Les histogrammes peuvent être définis comme une visualisation graphique des décomptes de données.
Il s'agit d'un type particulier de diagramme à barres.
Un histogramme montre la distribution d'une variable quantitative. Normalement, les diagrammes à barres sont utilisés pour comparer les variables.
Habituellement, il n'y a pas d'espace entre les colonnes pour les histogrammes.

Présentons un histogramme simple, en sélectionnant la colonne de puissance dans `mtcars`.

```r
qplot(mtcars$hp, geom="histogram")
```

Lorsque vous exécutez le code, vous pouvez recevoir le message d'erreur suivant :
`stat_bin()` en utilisant `bins = 30`. Nous pouvons choisir une meilleure valeur avec `binwidth`.
Binwidth définit la largeur de votre barre.
Améliorons encore notre histogramme, en changeant notre binwidth.

```r
qplot(mtcars$hp, geom="histogram", binwidth = 25)
```

Ok, maintenant chaque barre a une largeur de 25 pixels. Cependant, on peut voir que les visualisations de données entre 210 et 260 ch ne sont pas bien illustrées. Encore une fois, ajoutons un contour noir à l'histogramme pour le rendre plus facile à visualiser.

```r
qplot(mtcars$hp, geom="histogram", binwidth = 25, colour = I("black"))
```

Nous pouvons également définir des limites pour notre axe x. R ajustera automatiquement les meilleures valeurs selon nos données, donc le résultat ne changera pas lorsque vous utiliserez qplot et une limite adéquate.

```r
qplot(mtcars$hp, geom="histogram", binwidth = 25, colour = I("black"),xlim=c(50,350))
```

S'il arrive à "couper" une valeur, R affichera un message d'erreur indiquant combien de valeurs ne sont pas affichées.
Jouons avec les valeurs xlim, pour voir ce message. Vous pouvez mettre n'importe quelle valeur que vous souhaitez ici.

```r
qplot(mtcars$hp, geom="histogram", binwidth = 25, colour = I("black"),xlim=c(80,200))
```

Donnons maintenant un nom à nos axes. Nous utilisons les paramètres `xlab` et `ylab` pour les modifier.

```r
qplot(mtcars$hp, geom="histogram", binwidth = 25, colour = I("black"), xlab = "Horsepower", ylab= "Number of Cars")
```

Nous pouvons supprimer la couleur qui remplit nos barres en utilisant le paramètre alpha.

```r
qplot(mtcars$hp, geom="histogram", binwidth = 25, colour = I("black"), xlab = "Horsepower", ylab= "Number of Cars", alpha = I(0))
```

Comme nous l'avons fait avec le bar plot, nous pouvons donner un nom à notre graphique. Nous pouvons le faire en utilisant le paramètre `main`.

```r
qplot(mtcars$hp, geom="histogram", binwidth = 25, colour = I("black"), xlab = "Horsepower", ylab= "Number of Cars", alpha = I(0),
     main = "Histogram")
```

<h2 align=center>Diagrammes à secteurs</h2>

Un diagramme à secteurs est un graphique circulaire, le plus couramment utilisé dans les affaires. Il montre la proportion que chaque partie des données contribue à un total global. Cela se manifeste généralement comme une représentation du nombre de données qualitatives, comme les données démographiques.

Un diagramme à secteurs dans ggplot2 est un graphique à barres empilées transformé. Un graphique à barres empilées est un graphique qui empile toutes les valeurs sur l'axe vertical, au lieu de créer des barres séparées pour chaque point de données différent.
Commençons par créer un graphique à barres empilées.

```r
barp <- ggplot(mtcars, aes(x=1, y=sort(carb), fill=sort(carb))) +
        geom_bar(stat="identity")
print(barp)
```

Nous devons maintenant transformer ce graphique en diagramme à secteurs. Nous allons tracer notre graphique à barres empilées dans le système de coordonnées polaires.

Un très bon rappel sur les coordonnées polaires peut être vu dans le gif suivant :

<img src="https://ibm.box.com/shared/static/me1bmbb54gneyr5epweqe88zg8ujkqst.gif"</img>

Source: <a href="https://commons.wikimedia.org/wiki/File:Cartesian_to_polar.gif">Wikimedia Commons</a>

Pour ce faire, nous utilisons la fonction d'assistance `coord_polar` :

```r
barp <- barp + coord_polar(theta='y')

print(barp)
```

Nous pouvons utiliser la fonction d'assistance `theme` afin de ne pas déformer les étiquettes et les axes avec notre transformation :

```r
barp <- barp + coord_polar(theta='y')
barp <- barp + theme(
    axis.line=element_blank(),
    axis.text.x=element_blank(),
    axis.text.y=element_blank(),
    axis.ticks=element_blank(),
    axis.title.y=element_blank())
print(barp)
```

Vous pouvez remarquer que nous avons toujours un contour autour de notre graphique. Nous pouvons étendre la fonction de thème en utilisant le paramètre `panel.background` :

```r
barp <- barp + coord_polar(theta='y')
barp <- barp + theme(
    axis.line=element_blank(),
    axis.text.x=element_blank(),
    axis.text.y=element_blank(),
    axis.ticks=element_blank(),
    axis.title.y=element_blank(),
    panel.background=element_blank())
print(barp)
```

Il ne nous reste plus qu'à changer notre étiquette :

```r
barp <- barp + coord_polar(theta='y')
barp <- barp + theme(
    axis.line=element_blank(),
    axis.text.x=element_blank(),
    axis.text.y=element_blank(),
    axis.ticks=element_blank(),
    axis.title.y=element_blank(),
    panel.background=element_blank()) +
    labs(y="Carburetors")
print(barp)
```
